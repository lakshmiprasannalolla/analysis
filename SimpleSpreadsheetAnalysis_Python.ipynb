```python
#Importing Libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import datetime
import seaborn as sns 
%matplotlib inline 
```

```python
#Importing data
data = pd.read_excel(r'C:\Users\lakshmi lolla\Desktop\HireArt-Facebook\Dataset\HireArt - Data Analyst Exercise 10.12.17.xlsx')
```

```python
#Eliminating duplicates
data.drop_duplicates(inplace = True)
```

```python
#Formatting date string
year = []
month = []
day = []
for i in data.index:
        timestamp= str(data['Date of Contact'][i]).split()[0].split('-')
        year.append(timestamp[0])
        month.append(timestamp[1])
        day.append(timestamp[2])
```

```python
data['Year'] = year
data['Month'] = month
data['Day'] = day
```

```python
#Displaying how the first five records of the data is modified
data.head()
```

```python
#Eliminating specific duplicate values
data.drop_duplicates(subset=['Account manager','Client Name','Year','Month'],inplace=True)
```

```python
#Pivoting data
group = data.groupby(['Year','Month']).size() 
```

```python
#Exploratory data analysis using matplotlib 
#Comparing number of clients contacted vs year/month
plt.xlabel('Month')
plt.ylabel('Clients Contacted')
group.plot(kind='line', figsize=(10,10))
```

```python
group_df = pd.DataFrame(data=group)
months_sequence = [group_df.iloc[i].name[1]+'/'+group_df.iloc[i].name[0] for i in range(0,len(group_df))]
```

```python
#Exploratory data analysis using seaborn
#Comparing number of clients contacted vs month
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
plt.xlabel('Month')
plt.ylabel('Clients Contacted')
sns.barplot(x=months_sequence, y = group.values)
```
